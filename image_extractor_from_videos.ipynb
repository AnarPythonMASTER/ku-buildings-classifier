```python
VIDEOS_DIR = r"C:\Users\Shahbaz\Desktop\dl\videos"
FRAME_SKIP = 15 

print(f"Scanning {VIDEOS_DIR} for video files...\n")
for class_name in os.listdir(VIDEOS_DIR):
    class_path = os.path.join(VIDEOS_DIR, class_name)
    
    if not os.path.isdir(class_path):
        continue
        
    for video_file in os.listdir(class_path):
        if not video_file.lower().endswith(('.mp4', '.mov', '.avi')):
            continue
            
        video_path = os.path.join(class_path, video_file)
        video_name = os.path.splitext(video_file)[0] 
        
        # We removed the sub-folder creation. 
        # Everything will now save directly into the main class_path
        
        print(f"Processing: {class_name}/{video_file} -> extracting to /{class_name}/")
        cap = cv2.VideoCapture(video_path)
        
        frame_counter = 0
        saved_counter = 0
        
        while True:
            success, frame = cap.read()
            if not success:
                break
                
            if frame_counter % FRAME_SKIP == 0:
                # CRITICAL FIX: We add the video_name to the file to prevent overwriting
                frame_filename = f"{video_name}_frame_{saved_counter}.jpg"
                save_path = os.path.join(class_path, frame_filename)
                
                cv2.imwrite(save_path, frame)
                saved_counter += 1
                
            frame_counter += 1
            
        cap.release()
        print(f"  -> Extracted {saved_counter} frames from {video_file}.\n")

print("Video frame extraction complete!")
```
